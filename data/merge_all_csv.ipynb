```python
import pandas as pd
```

```python
df_arp = pd.read_csv('arp_poisoning_processed.csv')
df_ssh = pd.read_csv('dictionary_ssh_processed.csv')
df_normal = pd.read_csv('normal_processed.csv')
df_ddos = pd.read_csv('TCP_ddos_processed.csv')
```

```python
print(df_arp.shape)
print(df_ssh.shape)
print(df_normal.shape)
print(df_ddos.shape)
```

```python
df_arp['label'] = 'arp poisoning'
df_ssh['label'] = 'dictionary ssh'
df_normal['label'] = 'normal'
df_ddos['label'] = 'tcp ddos'
```

```python
df_final = pd.concat([df_arp,df_ssh,df_normal[:500_000],df_ddos[:500_000]]) 
```

```python
df_final.shape
```

```python
df_final.isnull().sum()
```

```python
tcp_mask = df_final['ip.proto'] == 6
udp_mask = df_final['ip.proto'] == 17
dns_mask = (df_final['udp.dstport'] == 53) | (df_final['udp.srcport'] == 53)

df_final.loc[~tcp_mask, ['tcp.*']] = 0
df_final.loc[~udp_mask, ['udp.*']] = 0
df_final.loc[~dns_mask, ['dns.*']] = 0
```

```python
df_final['label'].unique()
```

```python
df_final.to_csv('final.csv',index=False)
```
